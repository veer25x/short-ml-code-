```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\veerp\OneDrive\Certificate\Data set\Salary_Data.xls")
print(df.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=df[["YearsExperience"]]
y=df[["Salary"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
predection=model.predict(x_test)
print(predection)
```

```python
years = [[18]] 
predicted_salary = model.predict(years)
print("Predicted salary for 18 years experience:", predicted_salary[0])
```

```python
score = model.score(x_test, y_test)
print("R² Score:", score)
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
plt.scatter(x_train, y_train, color="blue", label="Training Data")
plt.plot(x_train, y_train_pred, color="red", label="Regression Line")
```

```python
plt.scatter(x_test, y_test, color="green", label="Test Data")

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression: Salary vs Experience")
plt.legend()
plt.show()
```
