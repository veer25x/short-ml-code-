```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y)          # Line plot
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Scatter Plot
plt.scatter(x, y)
# Pie Chart
sizes = [20, 30, 25, 25]
labels = ["A","B","C","D"]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
# Bar Plot
plt.bar(x, y)
```

```python
# Horizontal Bar
plt.barh(x, y)
```

```python
# Histogram
plt.hist([1,2,2,3,3,3,4,4,5], bins=5)
```

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=2)

plt.title("Styled Graph", fontsize=14, color='blue')
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)

plt.grid(True)     # Add grid
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [2,4,6,8,10]

plt.plot(x, y1, label="Square")
plt.plot(x, y2, label="Double")
plt.legend()    # Show labels
```

```python
import numpy as np

x = np.linspace(0, 10, 100)

plt.subplot(2,1,1)   # (rows, cols, position)
plt.plot(x, np.sin(x))
plt.title("Sine Wave")

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
plt.title("Cosine Wave")

plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)   # 100 points
y = np.sin(x)                 # also 100 points

plt.plot(x, y)
plt.savefig("graph.png")
plt.show()
```

```python
plt.plot(x, y, 'g--o')  # green, dashed, circle marker

plt.xlim(0, 6)    # X-axis range
plt.ylim(0, 12)   # Y-axis range

plt.xticks([1,2,3,4,5])  # Custom ticks
plt.yticks([2,4,6,8,10])
```
