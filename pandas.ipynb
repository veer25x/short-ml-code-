```python
#Pandas Data Structures
#Series (1D labeled array)
import pandas as pd

s = pd.Series([10, 20, 30, 40], index=['a','b','c','d'])
print(s)
```

```python
#DataFrame (2D table)
data = {
    'Name': ['Amit', 'Neha', 'Raj'],
    'Age': [22, 25, 28],
    'City': ['Delhi', 'Mumbai', 'Chennai']
}
df = pd.DataFrame(data)
print(df)
```
